#### 1.7 직접 데이터 프레임 생성

딕셔너리를 데이터프레임으로

```python
students = {
    'Name': ['Mike', 'Jane', 'Taylor', 'Post Malone'],
    'Gender': ['M', 'F', 'F', 'M'],
    'Major': ['Computer Science', 'Fine Arts', 'Mechanical Engineering', 'Economics'],
}
students_df = pd.DataFrame(students)
students_df
```

리스트를 데이터프레임으로

```python
students = [
    ['Mike', 'M', 'Computer Science'],
    ['Jane', 'F', 'Fine Arts'],
    ['Taylor', 'F', 'Mechanical Engineering'],
    ['Post Malone', 'M', 'Economics'],
]
column_name = ['Name', 'Gender', 'Major']
students_df = pd.DataFrame(students, columns=column_name)
students_df
```

원소가 딕셔너리인 리스트를 데이터프레임으로

```python
students = [
    {'Name': 'Mike', 'Gender': 'M', 'Major': 'Computer Science'},
    {'Name': 'Jane', 'Gender': 'F', 'Major': 'Fine Arts'},
    {'Name': 'Taylor', 'Gender': 'F', 'Major': 'Mechanical Engineering'},
    {'Name': 'Post Malone', 'Gender': 'M', 'Major': 'Economics'},
]
students_df = pd.DataFrame(students)
students_df
```

### 2. Feature 정보

#### 2.2 `State` feature 살펴보기

```python
customer['State'].value_counts()
```

#### 2.3 `State`이 `California`인 레코드만 보기

#### 2.4 `California`의 평균 `Income`

#### 2.5 `Education`별 `EmploymentStatus` 현황

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
%matplotlib inline

pd.crosstab(customer['EmploymentStatus'], customer['Education'], margins=True).style.background_gradient(cmap='coolwarm')
```

### 3. 새로운 feature 생성

#### 3.1 날짜 분리하기

```python
customer[['Month', 'Day', 'Year']] = customer['EffectiveToDate'].str.split('/', expand=True)
customer[['EffectiveToDate', 'Month', 'Day', 'Year']].head()
```

#### 3.2 `State`별 평균 `Income`

```python
customer['AverageIncome'] = customer.groupby(by='State')['Income'].transform('mean')
customer[['State', 'Income', 'AverageIncome']].head()
```

### 4. DataFrame 편집

#### 4.1 열 삭제

`axis=1`: 삭제할 대상이 0은 행, 1은 열

`inplace=True`: 데이터프레임에 변경사항 바로 적용


```python
customer.head()
```

```python
customer.drop('Customer', axis=1, inplace=True)
```

```python
customer.head()
```

#### 4.2 행 삭제

```python
customer.head()
```

```python
customer.drop(0, axis=0, inplace=True)  # 첫째행 삭제
```

```python
customer.head()
```

#### 4.3 index 번호 재설정

행을 삭제하면서 index 1번이 첫행이 되었으므로 0번 행이 되도록 index를 리셋해줌.

`drop=True`: index열을 삽입하지 않음

```python
customer.reset_index(drop=True, inplace=True)
```
